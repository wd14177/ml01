```python
from ast import literal_eval

result = literal_eval(df.iloc[0]['genres'])
result
```

```python
['Action','Adventure','Fantsy','Science Fiction']
```

```python
for temp in range(len(df)):
    df.iloc[temp]['genres'] = literal_eval(df.iloc[temp]['genres'])
```

```python
df['genres'] = df['genres'].apply(literal_eval)
```

```python
df.iloc[0]['genres']
```

```python
genres_list = []

for temp in range(len(df)):
    list_temp = []
    for temp1 in df.iloc[temp]['genres']:
        list_temp.append(temp1['name'])

    #list_temp = [Action, Adventure, Fantasy, Science Fiction]
    str_data = (' ').join(list_temp)
    #str_data = 'Action Adventure Fantasy Science Fiction'
    genres_list.append(str_data)
    #genres_list = ['Action Adventure Fantasy Science Fiction', 'Adventure Fantasy Action']


df['genres'] = genres_list
df['genres']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer(min_df=1, ngram_range=(1,2))
genre_mat = count_vec.fit_transform(df['genres'])
```

```python
count_vec.vocabulary_
```

```python
genre_mat.toarray()
```

```python
df_genre_vec = pd.DataFrame(genre_mat.toarray())
df_genre_vec.columns = sorted(count_vec.vocabulary_,key=count_vec.vocabulary_.get)
df_genre_vec.index = df['title']
df_genre_vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(df_genre_vec,df_genre_vec)
genre_sim
```

```python
genre_sim_df = pd.DataFrame(genre_sim)
genre_sim_df.columns= df['title']
genre_sim_df.index = df['title']
genre_sim_df
```

```python
index_num = df[df['title']=='Superman'].index.values[0]
genre_sim_df.iloc[index_num].sort_values(ascending=False)
```
