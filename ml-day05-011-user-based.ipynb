# User-based movie recommend system

```python
# C:\potenup3\ml01\data\movies.csv 열기
import pandas as pd
movies = pd.read_csv('data/movies.csv')
movies.head()
```

```python
movies.info()
```

```python
# C:\potenup3\ml01\data\ratings.csv 열기
ratings = pd.read_csv('data/ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
# ratings를 기반으로 movies data를 movieId 기준으로 병합
data = pd.merge(ratings, movies, on='movieId')
data.head()
```

```python
# genres 열의 문자열을 | 기준으로 분리하여 리스트로 변환
data['genres'] = data['genres'].str.split('|')
data.head()
```

```python
# userID 기준으로 pivot table 생성
user_movie_df = data.pivot_table('rating', index='userId', columns='title')
user_movie_df.head()
```

```python
# NaN 값을 0으로 변환
user_movie_df = user_movie_df.fillna(0)
user_movie_df.head()
```

```python
# 코사인 유사도 
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_movie_df)
user_similarity[:5]
```

```python
# df로 변환
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_df.index, columns=user_movie_df.index)
user_similarity_df.head()
```

```python
user_similarity_df.iloc[0].sort_values(ascending=False)
```

```python
# UserID 1번이 본 영화, 컬럼 값이 1인 컬럼 값 추출
user_movie_df.iloc[0][user_movie_df.iloc[0] > 0]

# userID 1번이 본 영화를 userID001 리스트로 저장
userID001 = user_movie_df.iloc[0][user_movie_df.iloc[0] > 0].index.tolist()
userID001
```

```python
print(len(userID001), "개의 영화를 봤습니다.")
```
