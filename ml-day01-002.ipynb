```python
# 레이블 인코더
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기', '냉장고', 'TV']

encoder = LabelEncoder()
encoder.fit(items)   # 통계량 학습
result = encoder.transform(items)  # 인코딩 수행
print('인코딩 변환값:', result)
```

```python
encoder.classes_  # 인코딩 대상 클래스 확인
```

```python
result = encoder.inverse_transform([3, 1, 2])  # 디코딩 수행
result
```

```python
# 데이터 가져오기
import pandas as pd
df = pd.read_csv('./data/internet_churn_rate.csv')
df.head()
df_raw = df.copy()  # 원본 데이터 복사본 생성
```

```python
# 라이브러리 가져오기
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# feature와 target 설정
feature = df[['월평균사용량_GB', '가입기간_년']]
target = df['이탈여부']  # target은 이탈여부 컬럼
```

```python
# feature 데이터 확인
feature.head()
```

```python
# target 데이터 확인
target.head()
```

```python
# label encoder 객체 생성
encoder = LabelEncoder()
# target 데이터 인코딩 변환
target_encoded = encoder.fit_transform(target)
target_encoded
```

```python
# 인코딩 변환된 클래스 확인
encoder.classes_
```

```python
# 디코딩 변환 수행
decoded = encoder.inverse_transform(target_encoded)
decoded
```

```python
# standard scaler 객체 생성
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

scaler = StandardScaler()
# feature 데이터 스케일링 변환
feature_scaled = scaler.fit_transform(feature)
feature_scaled
```
