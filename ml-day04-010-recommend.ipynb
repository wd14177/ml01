```python
# DATA CSV를 이용한 영화 추천 시스템 만들기
# C:\potenup3\ml01\data\tmdb_5000_movies.csv 에서 데이터를 불러오기
import pandas as pd
df = pd.read_csv('data/tmdb_5000_movies.csv')
df.head()
```

```python
# 몇몇 컬럼의 경우 내용이 너무 길어서 출력이 잘릴 수 있습니다.
pd.set_option('display.max_colwidth', 100)
df.head()
```

```python
# 리스트로 바꾸기
import ast
genres_list = ast.literal_eval(df.iloc[0]['genres'])
print(type(genres_list))
print(genres_list)
```

```python
# 반복문으로 뽑아오기
genre_names = []
for genre in genres_list:
    genre_names.append(genre['name'])
print(genre_names) # name 만 가져오기
```

```python
new_genres = [] # 모든 행에 대해서 장르 이름들만 뽑아서 리스트로 만들기
for x in df['genres']: # 각 행에 대해서
    parsed = ast.literal_eval(x) # 문자열을 리스트로 변환
    names = [] # 장르 이름들을 담을 리스트
    for genre in parsed: # 각 장르에 대해서
        names.append(genre['name']) # 장르 이름을 뽑아서 names에 추가
    new_genres.append(names) # new_genres에 추가

df['genres'] = new_genres
```

```python
# genres 컬럼의 각 행에 대해 ast.literal_eval을 사용하여 문자열을 리스트로 변환한 후, 
# 리스트 내의 각 딕셔너리에서 'name' 키의 값을 추출하여,
# 새로운 리스트로 반환하는 람다 함수를 적용
df['genres'] = df['genres'].apply(lambda x: [genre['name'] for genre in ast.literal_eval(x)]) 
```

```python
# 0 타이틀
df['title'][0]
```

```python
# 장르의 모든 행의 값을 하나로 묶는 집합 만들기
all_genres = set()
for genres in df['genres']:
    for genre in genres:
        all_genres.add(genre)
print(all_genres)
```

```python
# 집합을 리스트로 변환
all_genres = list(all_genres)
print(all_genres)
```

```python

# Genre_df 를 만들어서, index는 타이틀, columns는 장르명으로 만들기
genre_df = pd.DataFrame(0, index=df['title'], columns=all_genres) # 0으로 채우기
for i, genres in enumerate(df['genres']): # 각 행에 대해서
    for genre in genres: # 각 장르에 대해서
        genre_df.at[df['title'][i], genre] = 1 # 해당 장르에 1 넣기
genre_df.head()
```

```python
genre_df.sum().sort_values(ascending=False) # 장르별 영화 개수 세기
```

```python

# 영화 개수 순서대로 column 정렬
genre_df = genre_df[genre_df.sum().sort_values(ascending=False).index]
genre_df.head()
```

```python
# genre_df 를 csv 파일로 저장
genre_df.to_csv('data/genre_df.csv', index=True)
```

```python
# count vectorizer로 장르 컬럼을 벡터화하기
# 텍스트 데이터를 벡터화하는 도구
from sklearn.feature_extraction.text import CountVectorizer

# tokenizer는 각 행을 어떻게 토큰으로 나눌지 결정하는 함수, 
# lambda x: x는 각 행을 그대로 토큰으로 사용하겠다는 의미, 
# lowercase=False는 대소문자를 구분하겠다는 의미
count_vectorizer = CountVectorizer(tokenizer=lambda x: x, lowercase=False) 

# fit_transform은 데이터를 벡터화하는 과정, df['genres']는 장르 컬럼을 의미
genre_matrix = count_vectorizer.fit_transform(df['genres']) 

# (영화의 수, 장르의 수) 형태의 행렬이 만들어짐
print(genre_matrix.shape) 

# get_feature_names_out()은 벡터화된 행렬의 각 열이 어떤 장르를 나타내는지 알려주는 함수, 
# all_genres와 같은 순서로 나옴
print(count_vectorizer.get_feature_names_out()) 

# 벡터화된 행렬을 배열 형태로 출력, 각 행은 영화 하나를 나타내고, 
# 각 열은 장르 하나를 나타냄, 값이 1이면 해당 영화가 해당 장르에 속한다는 의미, 0이면 속하지 않는다는 의미
print(genre_matrix.toarray()) 

```

```python
# 코사인 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(genre_matrix, genre_matrix)
print(cosine_sim)
```

```python
# 코사인 유사도 데이터프레임 만들기
cosine_sim_df = pd.DataFrame(cosine_sim, index=df['title'], columns=df['title'])
cosine_sim_df
```

```python
# Avatar와 유사한 영화들 찾기
cosine_sim_df['Avatar'].sort_values(ascending=False)[1:20]
```
