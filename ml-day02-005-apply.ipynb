# scale 된 데이터 기반 grid search cv 적용

```python
# load dataset: train, test
# import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# load dataset
train_data = pd.read_csv('data/train_dataset.csv')
test_data = pd.read_csv('data/test_dataset.csv')
X_train = train_data.drop(columns=['Purchased'])
y_train = train_data['Purchased']
X_test = test_data.drop(columns=['Purchased'])
y_test = test_data['Purchased']
```

```python
train_data.head()
```

```python
# scale data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
# define model and parameter grid
rf = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [1, 10, 20],
    'min_samples_split': [2, 5]
}
```

```python
# setup grid search cv
grid_model = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
# fit model
grid_model.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_model.best_params_)
print("Best Score:", grid_model.best_score_)
```

```python
# predict and evaluate
y_pred = grid_model.best_estimator_.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
# best parameters
print("Best Parameters:", grid_model.best_params_)
```
